```python
# Predefined sets
a = {1, 2, 3, 4, 5}
b = {4, 5, 6, 7, 8}

print("Set A:", a)
print("Set B:", b)

print("\nChoose Operation:")
print("1. Union")
print("2. Intersection")
print("3. Difference (A - B)")
print("4. Symmetric Difference")

choice = int(input("Enter your choice (1-4): "))

match choice:
    case 1:
        s = a | b
        print("Union:", s)

    case 2:
        s = a & b
        print("Intersection:", s)

    case 3:
        s = a - b
        print("Difference (A - B):", s)

    case 4:
        s = a ^ b
        print("Symmetric Difference:", s)

    case _:
        print("Invalid choice")
```

```python
## We can perform CRUD (Create, Read, Update, Delete) operations on Dictionary
```

```python
 ## Using CRUD
 # -----------------------------
# DICTIONARY CRUD OPERATIONS
# -----------------------------

# Initial empty dictionary
students = {}

# -----------------------------
# 1. CREATE (Add data)
# -----------------------------

# Adding key-value pairs to dictionary
students["Anshu"] = 90
students["Rahul"] = 85
students["Priya"] = 92

# Printing dictionary after creation
print("After CREATE operation:")
print(students)


# -----------------------------
# 2. READ (Access data)
# -----------------------------

# Accessing value using key
print("\nREAD operation:")

# Print marks of a specific student
print("Marks of Anshu:", students["Anshu"])

# Loop through dictionary to read all data
print("All student marks:")
for name, marks in students.items():
    print(name, ":", marks)


# -----------------------------
# 3. UPDATE (Modify existing data)
# -----------------------------

print("\nUPDATE operation:")

# Updating Rahul's marks
students["Rahul"] = 88

# Showing updated dictionary
print("Updated dictionary:")
print(students)


# -----------------------------
# 4. DELETE (Remove data)
# -----------------------------

print("\nDELETE operation:")

# Removing Priya from dictionary
del students["Priya"]

# Printing dictionary after deletion
print("After deleting Priya:")
print(students)
```

```python
# Count the frequency of each value in the dictionary (for example, how many times "fruit" and "vegetable" appear).

data = {
    "apple": "fruit",
    "banana": "fruit",
    "carrot": "vegetable",
    "potato": "vegetable",
    "mango": "fruit",
    "cabbage": "vegetable",
    "orange": "fruit",
    "tomato": "vegetable",
    "grapes": "fruit",
    "onion": "vegetable",
    "pineapple": "fruit",
    "spinach": "vegetable"
}
d = {}
for value in data.values():
    if value in d:
        d[value] += 1
    else:
        d[value] = 1

print(d)
```

```python
# WAP to add the values of common keys and include the remaining key–value pairs from the second dictionary into the first dictionary.

d1 = {1:10, 2:20, 3:30, 4:40}
d2 = {4:44, 5:50, 6:60, 7:77}

for i in d2:
    if i in d1.keys():
        d1[i] += d2[i]
    else:
        d1[i] = d2[i]
print(d1)
```
