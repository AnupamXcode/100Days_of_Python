```python
# Example program demonstrating: try, except, else, finally, and raise
# Use case: A simple calculator that divides two numbers with proper error handling

try:
    # try block:
    # Used to place code that might cause an exception
    a = int(input("Enter numerator: "))
    b = int(input("Enter denominator: "))

    # raise:
    # Used to manually trigger an exception when a condition is not acceptable
    if b < 0:
        raise ValueError("Denominator should not be negative")

    result = a / b   # May cause ZeroDivisionError if b == 0

except ZeroDivisionError:
    # except block:
    # Handles specific exception when denominator becomes 0
    print("Error: Division by zero is not allowed.")

except ValueError as e:
    # Handles invalid input or manually raised ValueError
    print("Value Error:", e)

else:
    # else block:
    # Runs only if no exception occurs in the try block
    print("Division successful. Result =", result)

finally:
    # finally block:
    # Always executes whether an exception occurs or not
    # Common use: closing files, releasing resources, cleanup operations
    print("Program execution completed.")
```
