```python
# Create a function square(n) that returns square of number.
def square(n):
  squa=n*n
  print(squa)
square(4)
```

```python
# Create a function maximum(a, b) that returns the larger number.
def maximum(a,b):
  if (a>b):
    print(a)
  elif (a==b):
    print("Equal no.")
  else:
    print(b)
maximum(5,5)
```

```python
# Create a function sum_digits(n) that returns sum of digits.
def sum_digits(n):
  sum = 0
  while n>0:
    k = n % 10
    sum = sum + k
    n = n // 10
  return sum
sum_digits(123)
```
