```python
# Palindrome function

def Palindrome(st):
  rev = ""
  for i in range(len(st)-1,-1,-1):
    rev = rev + st[i]

  if rev == st:
    print("It's a Palindrome")
  else :
    print("Not a Palindrome")

Palindrome("naman")
Palindrome("amanaplanacanalpanama")
```

```python
# ================================
# PYTHON FUNCTIONS - COMPLETE GUIDE
# ================================

# -------- 1. Basic Function --------
def greet():
    print("Hello! Welcome to Python Functions.")

greet()   # Calling the function


# -------- 2. Function with Parameters --------
def greet_user(name):
    print("Hello,", name)

greet_user("Anshu")


# -------- 3. Function with Return Value --------
def add(a, b):
    result = a + b
    return result

sum_result = add(5, 3)
print("Sum =", sum_result)


# -------- 4. Default Parameters --------
def power(number, exponent=2):   # exponent default = 2
    return number ** exponent

print("Square:", power(4))
print("Cube:", power(4, 3))


# -------- 5. Keyword Arguments --------
def student(name, age):
    print("Name:", name)
    print("Age:", age)

student(age=20, name="Rahul")   # order doesn't matter


# -------- 6. Variable Length Arguments (*args) --------
# Used when number of inputs is unknown
def total_sum(*numbers):
    total = 0
    for num in numbers:
        total += num
    return total

print("Total:", total_sum(1, 2, 3, 4, 5))


# -------- 7. Keyword Variable Length Arguments (**kwargs) --------
def display_info(**data):
    for key, value in data.items():
        print(key, ":", value)

display_info(name="Anshu", age=21, city="Nagpur")


# -------- 8. Lambda Function (Anonymous Function) --------
# Small one-line function
square = lambda x: x * x
print("Lambda Square:", square(6))


# -------- 9. Recursive Function --------
# Function calling itself
def factorial(n):
    if n == 1:
        return 1
    return n * factorial(n - 1)

print("Factorial:", factorial(5))


# -------- 10. Nested Function --------
def outer_function(text):
    def inner_function():
        print("Inner Function says:", text)
    inner_function()

outer_function("Hello from Outer")


# -------- 11. Function with Docstring --------
def multiply(a, b):
    """
    This function returns multiplication of two numbers.
    Docstring explains what function does.
    """
    return a * b

print("Multiplication:", multiply(3, 4))
print(multiply.__doc__)   # Access docstring


# -------- 12. Global vs Local Scope --------
x = 10   # Global variable

def scope_demo():
    x = 5   # Local variable
    print("Inside Function x =", x)

scope_demo()
print("Outside Function x =", x)


# -------- 13. Using Global Keyword --------
count = 0

def increase():
    global count
    count += 1

increase()
increase()
print("Global Count:", count)


# -------- 14. Function Returning Multiple Values --------
def operations(a, b):
    return a + b, a - b, a * b

add, sub, mul = operations(10, 5)
print("Add:", add, "Sub:", sub, "Mul:", mul)


# -------- 15. Function as Argument --------
def apply_function(func, value):
    return func(value)

def double(x):
    return x * 2

print("Apply Function:", apply_function(double, 5))


# -------- END --------
print("\nYou have learned most Python function concepts 🎉")
```
